```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
koc=pd.read_csv('./KOC.csv')
koc.head()
%matplotlib inline
```

##Reading Data from WT.csv and KOC.csv

```python
wt=pd.read_csv('./WT.csv')
wt.head()
```

##Ploting the bar plot

```python
n_groups=6
fig,ax=plt.subplots()
index=np.arange(n_groups)
bar_width=0.15
opacity=0.4
error_config={'ecolor':'0.3'}
c=['#e60000',
'#ff8000',
'#ffff00',
'#00e666',
'#1ab3ff',
'#001acc',
'#b300ff']
l=koc.iloc[:,0]
for i in range(7):
    
      rects1 = plt.bar(index +i*bar_width, koc.iloc[i,(1,4,7,10,13,16)], bar_width,
                 alpha=opacity,
                 color=c[i],
                 yerr=koc.iloc[i,(2,5,8,11,14,17)],
                 error_kw=error_config,
                 label=l[i],
                      )
plt.xlabel('Time after initiation of conjugation (hr)')
plt.ylabel('Percentage of cells (%)')
plt.title('Time course analysis of progression of the nuclear events in delta CYC2')
plt.xticks(index + bar_width, ('2hr', '3hr', '4hr', '6hr', '10hr','24hr'))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
